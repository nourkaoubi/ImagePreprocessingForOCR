```python
import cv2
from matplotlib import pyplot as plt # type: ignore
image_file="data/letter.png"
img= cv2.imread(image_file)
#cv2.imshow("blabla", img)
cv2.waitKey(0)
def display_image_in_actual_size(im_path):

    dpi = 80
    im_data = plt.imread(im_path)
    height, width= im_data.shape

    # What size does the figure need to be in inches to fit the image?
    figsize = width / float(dpi), height / float(dpi)

    # Create a figure of the right size with one axes that takes up the full figure
    fig = plt.figure(figsize=figsize)
    ax = fig.add_axes([0, 0, 1, 1])

    # Hide spines, ticks, etc.
    ax.axis('off')

    # Display the image.
    ax.imshow(im_data, cmap='gray')

    plt.show()


#display_image_in_actual_size(image_file)


def grayscale(image):
    return cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
gray_image=grayscale(img)
cv2.imwrite("data/gray.jpg",gray_image)
display_image_in_actual_size("data/gray.jpg")

```

```python
thresh,im_bw=cv2.threshold(gray_image,200,230,cv2.THRESH_BINARY)
cv2.imwrite("data/bw_image.jpg",im_bw)
display_image_in_actual_size("data/bw_image.jpg")
```
